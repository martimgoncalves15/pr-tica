## Operadores para a construĂ§Ă£o de frases lĂ³gicas

Eis uma tabela de operadores que podem ser utilizados para formar proposiĂ§Ăµes. Note-se que temos um problema: queremos usar operadores Python para formar frases, para que os nossos programas (e as nossas sessĂµes interativas como esta aqui) mostrem cĂ³digo simples. Mas o Python nĂ£o permite setas de implicaĂ§Ă£o como operadores, pelo que, por enquanto, temos de usar uma notaĂ§Ă£o mais detalhada que o Python permite: `|'==>'|` em vez de apenas `==>`. Em alternativa, pode sempre utilizar as formas mais detalhadas do construtor `Expr`:

| OperaĂ§Ă£o | Livro | Entrada Infixa Python | SaĂ­da Python | Entrada Python `Expr`
|--------------------------|----------------------|-------------------------|---|---|
| NegaĂ§Ă£o | &not; P | `~P` | `~P` | `Expr('~', P)`
| E | P &and; P | `P & Q` | `P & Q` | `Expr('&', P, Q)`
| Ou | P &or; P | `P`<tt> &#124; </tt>`Q`| `P`<tt> &#124; </tt>`Q` | `Expr('`&#124;`', P, Q)`
| Desigualdade (Xor) | P &ne; P | `P ^ Q` | `P ^ Q` | `Expr('^', P, Q)`
| ImplicaĂ§Ă£o | P &rarr; P | `P` <tt>&#124;</tt>`'==>'`<tt>&#124;</tt> `Q` | `P ==> Q` | `Expr('==>', P, Q)`
| ImplicaĂ§Ă£o Reversa | P &larr; P | `Q` <tt>&#124;</tt>`'<=='`<tt>&#124;</tt> `P` |`Q <== P` | `Expr('<==', Q, P)`
| EquivalĂªncia | P &harr; P | `P` <tt>&#124;</tt>`'<=>'`<tt>&#124;</tt> `Q` |`P <=> Q` | `Expr('<=>', P, Q)`

Eis um exemplo de definiĂ§Ă£o de uma frase com uma seta de implicaĂ§Ă£o:

````python
~(P & Q)  |'==>'|  (~P | ~Q)
````

## `expr`: um atalho para construir frases

Se a notaĂ§Ă£o `|'==>'|` lhe parecer feia, pode utilizar a funĂ§Ă£o `expr`:

````python
expr('~(P & Q)  ==>  (~P | ~Q)')
````

`expr` recebe uma string como entrada e analisa-a num `Expr`. A string pode conter operadores de seta: `==>`, `<==` ou `<=>`, que sĂ£o manipulados como se fossem operadores infixos regulares do Python. E `expr` define automaticamente quaisquer sĂ­mbolos, pelo que nĂ£o precisa de os prĂ©-definir:

````python
expr('sqrt(b ** 2 - 4 * a * c)')
````

## Base de conhecimento `PropKB`

Vamos criar uma base de conhecimento, para isso vamos usar a classe `PropKB` que nos cria uma base de conhecimento vazia.

````python
kb = PropKB()
````

Vamos agora popular a nossa base de conhecimento com proposiĂ§Ăµes lĂ³gicas

````python
b = expr('~(P & Q)  ==>  (~P | ~Q)')
a = expr('sqrt(b ** 2 - 4 * a * c)')
c = associate('|',[expr('A'),expr('B')])
d = disjuncts(expr('~P | Q'))
e = conjuncts(expr('~C & D'))
f = to_cnf(expr('P & Q ==> Q'))
kb.tell(a)
kb.tell(b)
kb.tell(c)
kb.tell(d[0])
kb.tell(d[1])
kb.tell(~e[1])
kb.tell(f)
````

Podemos consultar o conteudo da base de dados

````python
kb.clauses
````

````python
kb.retract(a)
kb.clauses
````

````python
kb.ask(expr('~Q'))
````

````python
kb.ask_if_true(expr('P'))
````
